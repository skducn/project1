```python
# 逻辑运算 == & ~
# 获取性别为F的记录集
data[data["Gender"] == "F"]
```

```python
# 获取性别是女且Product_Category_2小于10的记录集
data[(data["Gender"] == "F") & (data["Product_Category_2"] < 10 )]
```

```python
# 获取性别是女且Product_Category_2小于1655的取反记录集
data[(data["Gender"] == "F") & ~(data["Product_Category_2"] < 1655 )]
```

```python
# 获取 Product_Category_2 值小于 10的记录集
data[(data["Product_Category_2"] < 10)]
```

```python
# 获取 Product_Category_2 为NaN的记录集
data[data["Product_Category_2"].isna()]
```

```python
# 获取 Product_Category_2 值为 2.0 和 8.0的记录集
data[data["Product_Category_2"].isin([2.0,8.0])]
```

```python
# apply 对多列或整个dataFrame处理数据类型
a = [['a', '1.2', '4.2'], ['b', '70', '0.03'], ['x', '5', '0']]
df = pd.DataFrame(a, columns=['col1','col2','col3'])
df
```

```python
df.dtypes
```

```python
df[['col2','col3']] = df[['col2','col3']].apply(pd.to_numeric)
df.dtypes
```

```python
# df.apply(pd.to_numeric, errors='ignore')
# 该函数将被应用于整个DataFrame，可以转换为数字类型的列将被转换，而不能(例如，它们包含非数字字符串或日期)的列将被单独保留。
```

```python
# infer_objects() 类型自动推断, 用于将具有对象数据类型的 DataFrame 的列转换为更具体的类型。
df = pd.DataFrame({'a': [7, 1, 5], 'b': ['3','2','1']}, dtype='object')
df.dtypes
```

```python
df
```

```python
df = df.infer_objects()
```

```python
df.dtypes
```

```python
# astype 强制转换
a = [['a', '1.2', '4.2'], ['b', '70', '0.03'], ['x', '5', '0']]
df = pd.DataFrame(a, columns=['one', 'two', 'three'])
df.dtypes
```

```python
df
```

```python
df[['two', 'three']] = df[['two', 'three']].astype(float)
df.dtypes
```
