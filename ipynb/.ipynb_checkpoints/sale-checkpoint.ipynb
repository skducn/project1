```python
import pandas as pd
import numpy as np
df = pd.read_excel("sales-funnel.xlsx")
df
```

```python
df["Status"] = df["Status"].astype("category")
df["Status"].cat.set_categories(["提交","待定","拒绝","成交"],inplace=True)
```

```python
df
```

```python
df.describe()
```

```python
pd.pivot_table(df,index=["Name"]).reset_index()
```

```python
pd.pivot_table(df,index=["Name","Rep","Manager"])
```

```python
pd.pivot_table(df,index=["Manager","Rep"])
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc="mean")
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"])
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=np.sum)
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=[np.mean,len])
```

```python
# aggfunc聚合函数只应用于values
# columns是对某一字段实际值进行统计，如，对不同产品
# 注意聚合函数结果是浮点数
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],columns=["Product"],aggfunc=["sum"]).reset_index()
```

```python
# fill_values = 0 表示对非数值NaN进行0填充， 如 fill_value =100 填充100
# 注意聚合函数结果是整数
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=["Product"],aggfunc=[np.sum],fill_value=0).reset_index()
```

```python
# 对价格和数量进行商品分割统计总数
pd.pivot_table(df,index=["Manager","Rep"],values=["Price","Quantity"],
               columns=["Product"],aggfunc=[np.sum],fill_value=0)
```

```python
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],aggfunc=[np.sum],fill_value=0)
```

```python
# margins=True 查看数据总和all
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
```

```python
pd.pivot_table(df,index=["Manager","Status"],values=["Price"],
               aggfunc=[np.sum],fill_value=0,margins=True)
```

```python
# 分别对price求和，对quantity求数
pd.pivot_table(df,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":np.sum},fill_value=0)
```

```python
table = pd.pivot_table(df,index=["Manager","Status"],columns=["Product"],values=["Quantity","Price"],
               aggfunc={"Quantity":len,"Price":[np.sum,np.mean]},fill_value=0)
table
```

```python
# 高级透视表过滤
# 一旦你生成了需要的数据，那么数据将存在于数据帧中。所以，你可以使用自定义的标准数据帧函数来对其进行过滤。
table.query('Manager == ["金浩"]')
```

```python
table.query('Status == ["待定","成交"]')
```
