```python
# 统计所有人一共花费多少钱

# 透视表(pivot_table) 对数据动态分布即分类汇总
# 特点：灵活性高，可随意指定你要的分析计算结果
# 操作性强，是一个报表生成神器
# 四个参数，index,values,columns,aggfunc
# index 是层次字段，"User_ID"第一层索引,"Gender"第二层索引
# values 统计计算的字段
# aggfunc 对数据集合进行的函数操作，缺省值为 mean(平均值)
gender_purchase = data.pivot_table(index=["User_ID","Gender"], values="Purchase",aggfunc="sum").reset_index().head(10)
gender_purchase
```

```python
data.pivot_table(index=["User_ID"],columns=["Product_ID"],values = ["Purchase"],aggfunc=["sum"],fill_value=0,margins=True).reset_index()
```

```python
gender_purchase = data.pivot_table(index=["Gender","User_ID"], values="Purchase",aggfunc="sum").reset_index()
gender_purchase
```

```python
gender_purchase.index
```

```python
gender_purchase["Purchase"][1]
```

```python
# 多少个消费者
gender_purchase.count()
```

```python
# 男性与女性购买占比
gender_count = gender_purchase.groupby(by="Gender").size().reset_index(name="人数")
gender_count
```

```python
gender_count["占比"] = gender_count["人数"]/gender_count["人数"].sum()
gender_count
```

```python
trace = go.Pie(labels=gender_purchase.Gender.tolist(),
              values = gender_purchase.Purchase.tolist(),
              hole=0.5
              )
layout = go.Layout(title="blackFriday男女消费对比")
fig = go.Figure(data=[trace],layout=layout)
fig.show()
```

```python
y_female = gender_purchase[gender_purchase.Gender == "F"].Purchase
y_male = gender_purchase[gender_purchase.Gender == "M"].Purchase
trace1 = go.Box(y=y_male,name="男性购物",boxmean=True)
trace2 = go.Box(y=y_female,name="女性购物",boxmean=True)
fig = go.Figure(data=[trace1,trace2],layout=layout)
fig.show()
```

```python
# 销量排名前10的商品
top10 = data.pivot_table(index=["Product_ID"],values="Purchase",aggfunc="count").reset_index().sort_values(by="Purchase",ascending=False).head(10)
top10
```

```python
# 那些人购买了商品？
# isin() 参数是列表，判断该列中元素是否存在。  element is In list()

top10.Product_ID
```

```python
top10.Product_ID.tolist()
```

```python
top_buyers = data[data.Product_ID.isin(top10.Product_ID.tolist())]
top_buyers
```

```python
# 每个商品男女购买力
top_seller_gender = top_buyers.pivot_table(index=["Product_ID","Gender"],values="Purchase",aggfunc="count").reset_index().head(10)
top_seller_gender
```

```python
traces=[]
i = 0
for g in top_seller_gender.Gender.unique():
    trace = go.Bar(x=top_seller_gender[top_seller_gender.Gender==g].Purchase,
                  y=top_seller_gender[top_seller_gender.Gender==g].Product_ID,
                  name=g,
                  orientation="h")
    traces.append(trace)
    i+=i
layout=go.Layout(title="热销产品销量与消费者性别的关系",width=800,height=500,margin=dict(l=200))
fig = go.Figure(data=traces,layout=layout)
fig.show()
```
