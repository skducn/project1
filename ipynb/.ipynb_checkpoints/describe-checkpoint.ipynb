```python
# 数据分析PANDAS之DATAFRAME.DESCRIBE() 用法概述
# 目的是观察这一系列数据的范围。大小、波动趋势

import pandas as pd
import numpy as np
dates = pd.date_range("20200115",periods=7)
dates
```

```python
df = pd.DataFrame(np.random.randn(7,5),index=dates,columns=list('ABCDa'))
df
```

```python
df.describe()
```

```python
# 参数1 percentiles，设定数值型特征的统计量，默认是[.25, .5, .75],也就是返回25%，50%，75%数据量时的数字
df.describe(percentiles=[.2,.75, .8])  # 默认有.5
```

```python
# 参数2 include ,describe()函数自动计算的字段有count（非空值数）、unique（唯一值数）、top（频数最高者）、freq（最高频数）
df = pd.read_excel("sales-funnel.xlsx")
df.describe(include=['O'])
```

```python
print(df.describe().iloc[3])   # 只获取min行数据
```
