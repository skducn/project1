##### 应用1：将三个表（df1,df2,df3）进行合并(result)

```python
df1 = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    },
    index=[0, 1, 2, 3],
)
```

```python
df1
```

```python
df2 = pd.DataFrame(
    {
        "A": ["A4", "A5", "A6", "A7"],
        "B": ["B4", "B5", "B6", "B7"],
        "C": ["C4", "C5", "C6", "C7"],
        "D": ["D4", "D5", "D6", "D7"],
    },
    index=[4, 5, 6, 7],
)
```

```python
df2
```

```python
df3 = pd.DataFrame(
    {
        "A": ["A8", "A9", "A10", "A11"],
        "B": ["B8", "B9", "B10", "B11"],
        "C": ["C8", "C9", "C10", "C11"],
        "D": ["D8", "D9", "D10", "D11"],
    },
    index=[8, 9, 10, 11],
)
```

```python
df3
```

```python
frames = [df1,df2,df3]
```

```python
result = pd.concat(frames)
```

```python
result
```

##### key用途1：分层索引

```python
# keys: 序列，默认无。使用传递的键作为最外层构建分层索引。如果通过了多个级别，则应包含元组。
result = pd.concat(frames, keys=["x", "y", "z"])
```

```python
result
```

```python
pieces = {"x": df1, "y": df2, "z": df3}
```

```python
result = pd.concat(pieces)
```

```python
result
```

```python
result = pd.concat(pieces, keys=["z", "y"])
```

```python
result
```

```python
result.index.levels
```

##### key用途2：对series列名覆盖

```python
s3 = pd.Series([0, 1, 2, 3], name="foo")
s4 = pd.Series([0, 1, 2, 3])
s5 = pd.Series([0, 1, 4, 5])
pd.concat([s3, s4, s5], axis=1)
```

```python
pd.concat([s3, s4, s5], axis=1, keys=["red", "blue", "yellow"])
```

```python
# 參考：https://pandas.pydata.org/pandas-docs/stable/user_guide/advanced.html#advanced-hierarchical
# 分层索引
result.loc["y"]
```

```python
result.loc["y"]["A"][5]
```

##### 两表交集(inner)与联合(outer)

```python
df4 = pd.DataFrame(
    {
        "B": ["B2", "B3", "B6", "B7"],
        "D": ["D2", "D3", "D6", "D7"],
        "F": ["F2", "F3", "F6", "F7"],
    },
    index=[2, 3, 6, 7],
)
```

```python
df4
```

```python
result = pd.concat([df1, df4], axis=1)  # result = pd.concat([df1, df4], axis=1, join="outer")
```

```python
result
```

```python
# join: {'inner', 'outer'}, 默认为 'outer'。如何处理其他轴上的索引。外部用于联合，内部用于交集。
result = pd.concat([df1, df4], axis=1, join="inner")
```

```python
result
```

```python
# 只获取df1的索引
result = pd.concat([df1, df4], axis=1).reindex(df1.index)
```

```python
result
```

##### 给 DataFrame 追加行记录

```python
# 1，通过Seris传递给append追加行
s2 = pd.Series(["X0", "X1", "X2", "X3"], index=["A", "B", "C", "D"])
```

```python
result = df1.append(s2, ignore_index=True)
```

```python
result
```

```python
# 2，通过dict传递给append追加行
dicts = [{"A": 1, "B": 2, "C": 3, "X": 4}, {"A": 5, "B": 6, "C": 7, "Y": 8}]
```

```python
result = df1.append(dicts, ignore_index=True)
```

```python
result
```
