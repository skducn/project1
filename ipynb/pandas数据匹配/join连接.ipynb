```python
import numpy as np
import pandas as pd
```

##### join()是将两个可能具有不同索引的列组合

```python
left = pd.DataFrame({"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"])
left
```

```python
right = pd.DataFrame(
    {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
)
right
```

```python
result = left.join(right)
result
```

```python
result = left.join(right, how="outer")  
# 等同于 result = pd.merge(left, right, left_index=True, right_index=True, how="outer")
result
```

```python
result = left.join(right, how="inner")
# 等同于 result = pd.merge(left, right, left_index=True, right_index=True, how="inner")
result
```

##### 连接索引上的键列¶


```python
left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "key": ["K0", "K1", "K0", "K1"],
    }
)
left
```

```python
right = pd.DataFrame({"C": ["C0", "C1"], "D": ["D0", "D1"]}, index=["K0", "K1"])
right
```

```python
result = left.join(right, on="key")
# 等同于 result = pd.merge(left, right, left_on="key", right_index=True, how="left", sort=False)
result
```

```python
result = pd.merge(left, right, left_on="key", right_index=True, how="left", sort=False)
result
```

##### 连接索引上的键列¶


```python
left = pd.DataFrame(
    {
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
    }
)
left
```

```python
index = pd.MultiIndex.from_tuples(
    [("K0", "K0"), ("K1", "K0"), ("K2", "K0"), ("K2", "K1")]
)
index
```

```python
right = pd.DataFrame(
    {"C": ["C0", "C1", "C2", "C3"], "D": ["D0", "D1", "D2", "D3"]}, index=index
)
right
```

```python
result = left.join(right, on=["key1", "key2"])
result
```

```python
result = left.join(right, on=["key1", "key2"], how="inner")
result
```

##### 将单个索引加入多索引¶


```python
left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]},
    index=pd.Index(["K0", "K1", "K2"], name="key"),
)
left
```

```python
index = pd.MultiIndex.from_tuples(
    [("K0", "Y0"), ("K1", "Y1"), ("K2", "Y2"), ("K2", "Y3")],
    names=["key", "Y"],
)
index
```

```python
right = pd.DataFrame(
    {"C": ["C0", "C1", "C2", "C3"], "D": ["D0", "D1", "D2", "D3"]},
    index=index,
)
right
```

```python
result = left.join(right, how="inner")
result
```

```python
# 等同于
result = pd.merge(
    left.reset_index(), right.reset_index(), on=["key"], how="inner"
).set_index(["key","Y"])
result
```
