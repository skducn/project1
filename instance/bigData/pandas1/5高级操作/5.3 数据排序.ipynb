```python
import pandas as pd
```

# 5.3.1 索引排序

```python
df = pd.read_excel('team.xlsx')
df.head()
```

```python
# 索引升序
df.sort_index().head()
```

```python
# 索引降序
df.sort_index(ascending=False).head()
```

```python
# 对列名进行降序
df.sort_index(ascending=False, axis=1).head()
```

```python
# 索引降序，输出team列值（Series）
df.team.sort_index(ascending=False).head()
```

```python
df.sort_index(ignore_index=True).head()
```

```python
# 空值在前， last表示空值在后
df.sort_index(na_position='first')
```

# 5.3.2 数值排序

```python
# 对Q1的值进行升序（Series）
df.Q1.sort_values().head()
```

```python
# 对team的值升序，再对name的值降序
df.sort_values(by=['team','name'],ascending=[True, False])
```
