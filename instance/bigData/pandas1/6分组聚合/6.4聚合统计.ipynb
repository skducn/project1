```python
import pandas as pd, numpy as np
```

# 6.4.1 描述统计

```python
df = pd.read_excel('../team2.xlsx')
df
                   
```

```python
# 描述统计
df.groupby('team').describe()
```

```python
df.groupby('team').describe().T
```

# 6.4.2 统计函数

```python
df.groupby('team').mean(numeric_only=True)
```

# 6.4.3 聚合方法 agg()

```python
df.groupby('team').sum(numeric_only=True)
# df.groupby('team').agg(sum)
# df.groupby('team').aggregate(sum)
```

```python
# 分组数量
df.groupby('team').agg(np.size)
```

```python
# 统计每组Q1的平均值
df.groupby('team').Q1.agg(np.mean)
```

```python
df
```

```python
# 分组统计Q1，Q3各自的总和与平均数
df.groupby('team')[['Q1','Q3']].agg([np.sum, np.mean])
```

```python
df.groupby('team').agg({'Q1':['min','max'],'Q2':'sum'})
```

```python
df.groupby('team').Q1.agg(Mean='mean', Sum='sum')
```

```python
# 分组team，对Q1求平均数，对Q2求总和
df.groupby('team').agg(Mean=('Q1','mean'), Sum=('Q2','sum'))
```
