```python
import pandas as pd
```

# 6.2.1 分组对象

```python
df = pd.read_excel("../team2.xlsx")

# 生成一个DataFrameGroupBy
df.groupby('team')
```

```python
# 生成一个SeriesGroupBy
df.Q1.groupby(df.team)
```

# 6.2.2 按标签分组

```python
# 按team分组，输出C组记录
df.groupby('team').get_group('C')
```

# 6.2.3 表达式

```python
# 按偶数索引分组求和，偶数为True，奇数False
df.groupby(df.index%2==0).sum(numeric_only=True)
df.groupby(lambda x:x%2==0).sum(numeric_only =True)  # 同上
```

```python
# 按条件索引分组，大雨5为True，索引0-4为False
df.groupby(df.index>4).sum(numeric_only =True)
# df.groupby(lambda x:x>4).sum(numeric_only =True) # 同上
```

```python
# 将列名包含Q的分成一组为True，其他为一组为False
df.groupby(lambda x:'Q' in x, axis=1).sum()
```

```python
df
```

```python
# 对名字首字母进行分组求和
df.groupby(df.name.str[0]).sum(numeric_only=True)
```

```python
# 按团队A和B为True，其他团队为False
df.groupby(df.team.isin(['A','B'])).sum(numeric_only=True)
```

```python
df
```

```python
# 按姓名第一个和第二个字母分组求和
df.groupby([df.name.str[0],df.name.str[1]]).sum(numeric_only=True)
```

# 6.2.4 函数分组

```python
# 按姓名首字母为元音、辅音分组
def getLetterType(l):
    if l[0].lower() in 'aeiou':
        return '元音'
    else:
        return '辅音'
df.set_index('name').groupby(getLetterType).sum(numeric_only=True)
```

```python
df
```

# 6.2.5 多种方法混合

```python
# 按team、姓名首字母是否为元音分组
df.groupby(['team',df.name.apply(getLetterType)]).sum(numeric_only=True)
```

# 6.2.6 用pip调用分组方法

```python
pd.DataFrame.groupby
```

```python
df.pipe(pd.DataFrame.groupby, 'team').sum(numeric_only=True)
```

# 6.2.7 分组器Grouper

```python
df.groupby(pd.Grouper('team'))
```

```python
df.groupby('team')
```

```python
df.groupby(pd.Grouper('team')).sum(numeric_only=True)
# df.groupby('team').sum(numeric_only=True) 同上
```

# 6.2.8 　索引

```python
# 对team分组，且不让team成为索引。
df.groupby('team', as_index=False).sum(numeric_only=True)
```

# 6.2.9 排序

```python
# 索引排序
df.groupby('team', sort=True).sum(numeric_only=True)
```

```python
# 索引不排序
df.groupby('team', sort=False).sum(numeric_only=True)
```
