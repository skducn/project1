# 6.1.3 应用

```python
df = pd.read_excel("../team2.xlsx")
df
```

```python
# 对team进行分组，对应列相加
df.groupby('team').sum(numeric_only=False)
```

```python
# 分组聚合，对不同列使用不同计算方法
df.groupby('team').agg({'Q1':sum,'Q2':'count','Q3':'mean','Q4':max})
```

```python
# 对同一列使用不同计算方法：
df.groupby('team').agg({'Q1':[sum,'std',max],'Q2':'count','Q3':'mean','Q4':max})
```

# 6.1.4 Series应用分组

```python
# 对team分组获取Q1的总数
df.Q1.groupby(df.team).sum()
```
