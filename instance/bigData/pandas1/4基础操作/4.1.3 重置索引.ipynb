# 4.1.3 重置索引 reset_index()

```python
df = pd.read_excel('team.xlsx',index_col='name')
df.head()
```

```python
# 1，重置索引
df.reset_index(inplace=True)
df.head()
```

```python
# 2，设置索引并取消索引，相当于什么都没做。
df = df.set_index('name').reset_index()
df.head()
```

```python
# 3, 应用：通过设置索引name，再删除索引且不保留原数据，达到删除原name列的作用。
df = df.set_index('name').reset_index(drop=True)
df.head()
```

```python
# 4，重置一级（level=1）索引，即['team','name']
df = pd.read_excel('team.xlsx')
df.set_index(['team','name']).reset_index(level=1,inplace=True)
df.head()
```

```python
df.set_index(['Q1','Q3'],inplace=True)
df.head()
```

```python
# 5，重置Q3索引
df.reset_index(level='Q3',inplace=True)
df.head()
```

```python
df.reset_index(col_level=0,inplace=True)
df.head()
```

```python
df.set_index(['Q1','Q2','Q4'],inplace=True)
df.head()
```

```python
df.reset_index(level='Q1',col_level=0,inplace=True)
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```
