# 4.1.7 索引的操作

```python
df = pd.read_excel('team.xlsx') 
df.head()
```

```python
# 【统计】
# 1，统计索引不重复的数量
df.index.nunique()
```

```python
# 【类型】
# 1，获取索引的类型
df.index.dtype
```

```python
# 2 将索引转换为浮点型
df.index = df.index.astype('float64')
```

```python
df.head()
```

```python
# 【操作名】
# 1,修改索引名
df.index = df.index.rename("haha")
df.head()
```

```python
# 2，表头加前缀
df.add_prefix('t_').head()
```

```python
# 3，表头加后缀 
df.add_suffix("_d").head()
```

```python
df.index.rename("hello")
```

```python
# [操作值]
# 1，索引倒序
df.index = df.index.sort_values(ascending=False)
```

```python
df.head()
```

```python
# 2，索引计算
df.index = df.index + 10
df.head()
```

```python
# 3，索引转字符串并添加后缀
df.index = df.index.map(lambda x:str(x) + "_")
df.head()
```

```python
# 4, 使用map函数去掉后缀_
df.index = df.index.map(lambda x:x[:-1])
df.head()
```

```python
# 5，索引字符串替换
df.index = df.index.str.replace(".","")
df.head()
```

```python
# 8，字符分割
df.index = df.index.str.split("10")
df.head()
```

```python
# 取消索引
df.reset_index(inplace=True)
df.head()
```

```python
# [剥离索引值]
# 1，剥离索引值称为另一个DF
df.index.to_frame(index=False, name='a').head()
```

```python
# 2，剥离索引值成为一个列表
df.index.to_list()
```

```python
df.index.to_list()[2]
```

```python
# 3,剥离索引值成为一个series
df.index.to_series()
```

```python
# 4, 剥离索引值成为一个numpy 
df.index.to_numpy()
```

```python
# 16, 去重及计数
x = df.index.value_counts()
x
```

```python
# 17，筛选条件,不满足的改诶NaN
df.index = df.index.where(df.index==11)
```

```python
# 18，索引最大值
df.index.max()
```

```python
# 19，最大索引号
df.index.argmax()
```

```python
# 20，转置，在多层索引里有用
df.index.T
```

```python
# 21，有没有空值
df.index.hasnans
```

```python
# 22，推测数据类型
df.index.inferred_type
```

```python
# 23，重复2次
df.index.repeat(2)
```

```python
# 24，索引层级数，通常为1
df.index.nlevels
```
