 # 4.4 位置计算

```python
import pandas as pd
```

```python
df = pd.read_excel('team2.xlsx',index_col='name')
# df = pd.read_excel('team2.xlsx',index_col=[0,1])
df
```

# df位置偏移计算(数字列)

```python
# 1, 如：计算每个学生每个季度较前一个季度成绩的变化值，即后一个季度成绩减去上一个季度成绩，第一季度Q1忽略。
df.loc[:,'Q1':'Q4'].diff(1,axis=1)
```

# series位置偏移计算

```python
# diff 后减前，默认移动1，第一位9变NaN，4-9，6-2，7-6，9-7
d = pd.Series([9,4,6,7,9])
d.diff()
```

```python
d.diff()[3]
```

```python
# diff 后减前，移动2 ，即，9和4为NaN，6-9，7-4，9-6
d.diff(2) # 向左移动2位
```

```python
# diff 后减前，移动-2，即，9-6，4-7，6-9，7和9为NaN
d.diff(-2)  # 向右移动2位
```

# 4.4.2 位置移动 shift（）

```python
df = pd.read_excel('team2.xlsx',index_col='name')
# df = pd.read_excel('team2.xlsx',index_col=[0,1])
df
```

```python
# 整体下移1行，第一行为NaN
df.shift(1)
```

```python
# 整体右移1列
df.shift(axis=1)
```

```python
# 整体左移1列
df.shift(-1,axis=1)
```

```python
# Q1列上移1行，最后一行为NaN
df.Q1.shift(-1)
```

```python
# Q1列下移1行
ff=df.Q1.shift()
ff
```

```python
# 将ff存入字典
dict={}
for index, value in ff.items():
    dict[index] = value
dict
```

```python
# 获取数据类型
type(df.Q1.shift()['Ack'])
```

```python
type(df.Q1.shift()['Ack'])
for index, row in df.iterrows():
    print(f'Index: {index}, Row: {row}')
```

```python
df.Q1 - df.Q1.shift()
```
