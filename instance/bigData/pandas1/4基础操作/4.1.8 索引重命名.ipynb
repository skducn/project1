# 4.1.8 索引重命名

```python
df = pd.read_excel('teamall.xlsx',index_col='name')
df.head()
```

```python
# 设置多层索引
df.set_index('Q4',append=True,inplace=True)
df.head()
```

```python
# 1，修改索引名(列表方式)
df.rename_axis(['baidu','dy']).head()
```

```python
# 2，修改索引名（字典方式）
df.rename_axis(index={'name':'A','Q4':'B'}).head()
```

```python
# 3，修改列名
df.rename(columns={'team':'kk','Q1':'BB'}).head()
```

```python
# 4, 将列名改为小写
df.rename(str.lower, axis='columns').head()
```

```python
# 5, 修改部分行索引值
df.rename(index={7:"x",'Leo':678})
# df.rename({7:"x",'Leo':678}, axis='index') # 同上
```

```python
# 修改数据类型，即 name和Q4转为字符串
# df = df.rename(index=str)
```

```python
df = df.reset_index('Q4').set_index('Q3',append=True)
df.head()
```

```python
df.loc[:,'Q4']['Lfie'][99]
```

```python
# 6,修改列名加前缀t_
df.rename(lambda x:'t_' + x, axis=1).head()
```

```python
# 7,修改全部列名
df.set_axis(['Qteam','QQ1','QQ2','QQ3'],axis='columns').head()
```

```python
# 8，修改全部列名
df.set_axis(list('abcd'), axis=1).head()
```

```python
# 9,修改索引值
df.set_axis(pd.Index(list('123456789b')), axis=0)
```
