# 4.4.1 位置差值 diff()
# 用于数字计算，一般用于增量研究

```python
d = pd.Series([9,4,6,7,9])
```

```python
# 第一位由于被减数下移，没有数据，所以结果为NaN
# 即，9变NaN，4-9，6-2，7-6，9-7
d.diff()
```

```python
d.diff()[3]
```

```python
# 参数-2表示移动多少位
# 即，9和4为NaN，6-9，7-4，9-6
d.diff(2)
```

```python
# 负数代表移动方向相反
# 即，9-6，4-7，6-9，7和9为NaN
d.diff(-2)
```

# 4.4.2 位置移动 shift（）

```python
df = pd.read_excel('team2.xlsx',index_col='name')
# df = pd.read_excel('team2.xlsx',index_col=[0,1])
df
```

```python
# 整体下移，第一行为NaN
df.shift(1)
```

```python
# Q1列上移一行，最后一行为NaN
df.Q1.shift(-1)
```

```python
# 向右移动1位
df.shift(axis=1)
```

```python
# 向左移动1位
df.shift(-1,axis=1)
```

```python
df.Q1
```

```python
df.Q1.shift()
```

```python
df.Q1 - df.Q1.shift()
```
