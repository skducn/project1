# 5.4.6 插入列 df.insert()

```python
df
```

```python
# 在第三列插入总和，值为sum(1)
df.insert(2,'总和', df.sum(1,numeric_only=True))
df
```

# 5.4.7 指定列df.assign()

```python
df = pd.read_excel('team2.xlsx')
df
```

```python
# 链式方式，total是虚拟的，
df.assign(total=df.sum(1,numeric_only=True))
```

```python
df
```

```python
df.assign(total=df.sum(1,numeric_only=True),Q=100)
# df.assign(total=df.sum(1)).assign(Q=100)  # 同上
```

```python
# avg2计算用到了total，但实际total列不存在，只能用lambda方式调用。
(df.assign(total=df.sum(1,numeric_only=True)).assign(Q=100).assign(name_len=df.name.str.len()).assign(avg=df.mean(1,numeric_only=True)).assign(avg2=lambda d: d.total/4))
```

```python
df.assign(Q6=df.Q2/df.Q1)
```

```python
# 逻辑表达式结果True，False
df.assign(tag=df.Q1>df.Q2)
```

```python
# 比较计算，True为1，False为0
df.assign(tag=(df.Q1>df.Q2).astype(int))
```

```python
# map映射文案
df.assign(tag=(df.Q1>60).map({True:"合格",False:"不及格"}))
```

```python
# 增加多个值，特别是Q9中由于Q8不是实列，只能通过lambda调用
df.assign(Q8=lambda d: d.Q1*5, Q9=lambda d: d.Q8+1)
```

# 5.4.8 执行表达式

```python
df = pd.read_excel('team2.xlsx')
# total是虚拟的
df.eval('total=Q1+Q2+Q3+Q4')
```

```python
# 赋值给了C1，df实体中存了C1
df['C1'] = df.eval('Q2 + Q3')
df
```

```python
# C2是虚拟的
df.eval('C2 = Q2 + Q3')
```

```python
# 通过变量
a= df.Q1.mean()
df.eval('C3 = `Q3` + @a')
```

```python
# 获取C3布尔值，先`Q3` + @a ，再与Q2逻辑比较
df.eval('C3 = Q2 > `Q3` + @a')
```

```python
# C4通过inplace=True变为实体列
df.eval('C4 = name + team', inplace=True)
df
```

# 5.4.10  追加合并

```python
df = pd.DataFrame([[1,2],[3,4]], columns=list('AB'))
df2 = pd.DataFrame([[5,6],[7,8]], columns=list('AB'))
# df.append(df2)  # 后续废弃
pd.concat([df,df2], ignore_index=True)
```

```python
s1 = pd.Series(['a', 'b'])
s2 = pd.Series(['c', 'd'])
pd.concat([s1, s2])
```

```python
# 索引重新编
pd.concat([s1, s2], ignore_index=True)
```

```python
# 多层索引，i1，i2
pd.concat([s1, s2],keys=['i1','i2'])
```

```python
a = pd.concat([s1, s2],keys=['i1','i2'])
a
```

```python
a['i2'][1]
```

```python
# 带列名的Series
pd.concat([s1, s2],keys=['i1','i2'], names=['Series name','Row Id'])
```
