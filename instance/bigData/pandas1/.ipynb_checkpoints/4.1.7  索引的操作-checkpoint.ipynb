```python
import pandas as pd
import numpy as np
```

# 4.1.7 索引的操作

```python
df = pd.read_excel('team.xlsx') 
df.head()
```

```python
# 1，转换类型 astype()
df.index.astype('int64')
```

```python
# 1，转换索引为浮点型
df.index = df.index.astype('float64')
```

```python
df.head()
```

```python
# 2，修改索引名称
df.index = df.index.rename("haha")
df.head()
```

```python
df.index.rename("hello")
```

```python
# 3，不重复的数量
df.index.nunique()
```

```python
# 4，排序
# 进行倒序
df.index = df.index.sort_values(ascending=False)
```

```python
df.head()
```

```python
# 5，索引计算
df.index = df.index + 10
df.head()
```

```python
# 6, map函数
# 索引改为字符串且添加后缀_
df.index = df.index.map(lambda x:str(x) + "_")
df.head()
```

```python
# 6，使用map函数去掉后缀_
df.index = df.index.map(lambda x:x[:-1])
df.head()
```

```python
# 7，字符串替换
df.index = df.index.str.replace(".","")
df.head()
```

```python
# 8，字符分割
df.index = df.index.str.split("10")
df.head()
```

```python
# 9，取消索引，索引变为列
df.reset_index(inplace=True)
df.head()
```

```python
# 10，索引计算
df.index = df.index + 10
df.head()
```

```python
# 11，转换成dataframe
df.index.to_frame(index=False, name='a').head()
```

```python
# 12，索引转为列表
df.index.to_list()
```

```python
df.index.to_list()[2]
```

```python
# 13, 转为Series
df.index.to_series()
```

```python
# 14, 转为numpy
df.index.to_numpy()
```

```python
# 15，去重
df.index.unique()
```

```python
# 16, 去重及计数
x = df.index.value_counts()
```

```python
# 17，筛选条件
df.index = df.index.where(df.index==11)
```

```python
df
```

```python
# 18，索引最大值
df.index.max()
```

```python
# 19，最大索引号
df.index.argmax()
```

```python
# 20，转置，在多层索引里有用
df.index.T
```

```python
# 21，有没有空值
df.index.hasnans
```

```python
# 22，推测数据类型
df.index.inferred_type
```

```python
# 23，重复2次
df.index.repeat(2)
```

```python
# 24，索引层级数，通常为1
df.index.nlevels
```

```python
# 表头加后缀
df.add_suffix('_d')
```
