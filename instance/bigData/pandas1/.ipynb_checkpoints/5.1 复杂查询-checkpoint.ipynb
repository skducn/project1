5.1.2 切片逻辑筛选数据

```python
# 2，索引array类数组逻辑运算（索引为1的值为True，其他为False）
df.index == 1
```

```python
# 3，DataFrame逻辑运算（满足Q1到Q4所有行大于50为True，否则为False）
df.loc[:,'Q1':'Q4']>50
```

```python
# 4 组合条件位运算 (Q1成绩不小于60分，且是C成员)
~(df.Q1 <60) & (df['team'] == 'C')
```

# 5.1.2 逻辑筛选数据

```python
# 1,Q1中等于93的行数据
df[df['Q1']==93]
```

```python
# 2,Q1中不等于93的行数据
df[~(df['Q1']==93)]
```

```python
df[df.name=='Ack']
# df[df['name']=='Ack'] # 同上
```

```python
df[df.Q1>df.Q2]
```

```python
# 获取符合条件一列的Series组合
df.loc[df['Q1']>90 , 'Q1']
```

```python
# 获取符合条件的多列DataFrame组合
df.loc[(df['Q1']>90) & (df.Q2 <15)]
```

```python
df.loc[df.Q1 ==89]
```

```python
df.loc[df.Q1>90, 'Q1':]
```

```python
# 逻辑运算
df.loc[:,['Q1','Q2']] > 50
```

```python
a = df.loc[:,'Q1':'Q2']>50
# Q1和Q2值都满足时，all返回True
df[a.all(1)]
```

```python
# Q1和Q2值两者有一个满足时，any返回True
df[a.any(1)]
```

```python
df.loc[:,['Q1','Q2']] > 50
```

```python
# 将不符合条件的值改为NaN
df[df.loc[:,['Q1','Q2']] > 50]
```

# 5.1.3 函数筛选

```python
df
```

```python
# 获取索引的最大值
max(df.index)
```

```python
# 获取索引4的Q1 Series
df.Q1[df.index==4]
```

```python
# 获取最大索引的Q1值
df.Q1[max(df.index)]
```

```python
df[lambda df: df['Q1'] == 93]
```

```python
# 前三行，显示列名长度为2的列
df.iloc[:3, lambda df: df.columns.str.len()==2]
```

```python
# 同上，前三行，列名中包含Q的列
df.loc[:2, lambda df : [i for i in df.columns if 'Q' in i ]]
```

# 5.1.4 比较函数

```python
df[df.Q1.eq(89)]
# 等同于 df[df.Q1 == 89]
```

```python
# df.ne() 不等于 ！=
# df.le() 小于等于 <=
# df.lt() 小于 <
# df.ge() 大于等于 >=
# df.gt() 大于 >
```

```python
df
```

```python
# 判断isin是否包含指定内容
df[df.team.isin(['A','B'])]
```

```python
df[df.name.isin(['Ack','B'])]
```

```python
# 复杂查询，其他值为NaN
df[df.isin({'item':['C','D'],'Q1':[36,93]})]
```

# 5.1.6 筛选df.filter()

```python
# 显示Q3、Q1列
df.filter(items=['Q3','Q1'])
```

```python
# 正则，显示包含Q列
df.filter(regex='Q', axis=1)
```

```python
# 正则，列名最后时e的列
df.filter(regex='e$', axis=1)
```

```python
# 正则索引名以1结尾的行
df.filter(regex='1$', axis=0)
```

```python
# 正则索引中有2的
df.filter(like='1', axis=0)
```

```python
# 正则索引中以2开头，列名中有Q的
df.filter(regex='^2', axis=0).filter(like='Q', axis=1)
```
