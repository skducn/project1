```python
import pandas as pd
```

# 4.1.3 重置索引 reset_index()

```python
df = pd.read_excel('team.xlsx',index_col='name')
df.head()
```

```python
# 1，消除索引
df.reset_index(inplace=True)
df.head()
```

```python
# 2，相当于什么都没做
df.set_index('team').reset_index()
df.head()
```

```python
# 设置索引name，删除原索引，name列没了
df.set_index('name').reset_index(drop=True)
df.head()
```

```python
# 一级索引取消
df.set_index(['team','name']).reset_index(level=1,inplace=True)
df.head()
```

```python
# 取消索引Q2，注意需要加上drop=True
df.set_index(['Q1','Q3'],inplace=True)
df.head()
```

```python
# 使用层级索引名取消索引
df.reset_index(level='Q3',inplace=True)
df.head()
```

```python
df.reset_index(level='Q1',col_level=1,inplace=True)
df.head()
```

```python
df.set_index(['Q1','Q2','Q4'],inplace=True)
df.head()
```

```python
df.reset_index(level='Q1',col_level=1,inplace=True)
df.head()
```

```python
df.reset_index(level='Q2',col_level=1, col_fill='species',inplace=True)
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
# 调整列名顺序方法1，手工排序
df.head()[['name','team','Q1','Q2','Q3','Q4']]
```

```python
# 调整列名顺序方法2, loc[]
df.head().loc[:,['name','team','Q1','Q2','Q4','Q3']]
```

```python
# 调整列名顺序方法3,iloc自然索引
df.iloc[:,[0,2,3,4,1,5]]
df.head()
```

```python
# 调整列名顺序方法4,逆序1
cols = list(df.columns)
cols.reverse()
df[cols]
```

```python
# 调整列名顺序方法4,逆序2
cols = list(df.columns)
df[cols[::-1]]
```
