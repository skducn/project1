```python
import pandas as pd, numpy as np
```

```python
df = pd.read_excel("../team.xlsx")
df.head()
```

```python
# 添加标题
df.head().style.set_caption('学生成绩表')
```

```python
# 全局数据精度，即保留2位小数
df.head().style.format(precision=2)
# df.head().style.set_precision(2) # 同上，即将弃用
```

```python
df.head().round(2).style
```

```python

df.assign(mean=df.mean(1, numeric_only=True)).head().style.format(precision=4)
# df.assign(mean=df.mean(1, numeric_only=True)).head().style.set_precision(4) # 同上，即将弃用
```

```python
# na_rep设置缺失值默认值
# 将Q4设为nan
na = np.nan
df.head().eval('Q4=@na').style.format(na_rep='暂无')
# df.head().eval('Q4=@na').style.set_na_rep('暂无')
```

```python
# 不输出指定列
df.head().style.hide('Q3',axis='columns') 
# df.head().style.hide_columns(['Q3']) # 同上，即将弃用。
```

# 15.3.2 表格CSS样式

```python
# 将Q1值改为红色
df.head().style.set_properties(subset=['Q1'], **{'color':'red'})
```

# 15.3.3 应用函数

```python
# applymap()对全表起作用，将大于30分的格子设置黄色背景。
bg = lambda x: 'background-color:yellow' if type(x) == int and x>30 else ''
df.head().style.applymap(bg)
```

```python
# 对指定列，将大于50分的格子设置黄色背景
df.style.applymap(lambda x:'background-color:yellow' if x>50 else '', subset=pd.IndexSlice[:,['Q1','Q2']])
```
