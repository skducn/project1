# 综合运用

```python
# 多个format无法连用？
df.head(10).assign(avg=df.mean(axis=1, numeric_only=True)/100) \
.assign(diff=lambda x: x.avg.diff()) \
.style \
.format({'1city':str.upper}).format({'diff':'¥{:.2f}'}, na_rep='-',).format({'avg': '{:.2%}'})
```
